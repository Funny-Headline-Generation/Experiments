Harvard

```python
with open(base_folder+'Harvard/NewsArticles.csv', encoding='utf-8', errors='ignore') as inp:
    harvard = pd.read_csv(inp)
```

```python
harvard.head()
```

```python
harvard['content'] = harvard['text']
harvard['id'] = harvard['article_id']
```

```python
harvard['source'] = 'Harvard'
times['source'] = 'Times'
components_one['source'] = 'AllTheNewsComponentsOne'
kaggle['source'] = 'AllTheNewsKaggle'

news_df = pd.concat([harvard[['id','title','content','source']],
                    times[['id','title','content','source']],
                    components_one[['id','title','content','source']],
                    kaggle[['id','title','content','source']]],
                   ignore_index=True)
```
